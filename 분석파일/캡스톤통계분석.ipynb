```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Excel 파일 로드
#df = pd.read_excel('your_data_file.xlsx', sheet_name='Sheet1')

# 고객 만족도 점수 계산 (예제 열 이름 사용)
#df['customer_satisfaction_score'] = df[['satisfaction_q1', 'satisfaction_q2', 'satisfaction_q3', 'satisfaction_q4']].mean(axis=1)

def anova_analysis(dependent_var, independent_var, df):
    model = ols(f'{dependent_var} ~ C({independent_var})', data=df).fit()
    anova_results = sm.stats.anova_lm(model, typ=2)
    return anova_results

# 분석 함수
def perform_anova(df, dependent_var, independent_var):
    anova_results = anova_analysis(dependent_var, independent_var, df)
    means = df.groupby(independent_var)[dependent_var].mean()
    stds = df.groupby(independent_var)[dependent_var].std()
    counts = df.groupby(independent_var)[dependent_var].count()
    total_mean = df[dependent_var].mean()
    total_std = df[dependent_var].std()
    total_count = df[dependent_var].count()

    results = []
    for group in counts.index:
        results.append({
            'Dependent Variable': dependent_var,
            'Independent Variable': independent_var,
            'Group': group,
            'N': counts[group],
            'Mean': means[group],
            'Std Dev': stds[group],
            'F-value': anova_results['F'][0],
            'p-value': anova_results['PR(>F)'][0]
        })

    # 전체 평균 및 표준편차 추가
    results.append({
        'Dependent Variable': dependent_var,
        'Independent Variable': independent_var,
        'Group': 'Overall',
        'N': total_count,
        'Mean': total_mean,
        'Std Dev': total_std,
        'F-value': '',
        'p-value': ''
    })

    results_df = pd.DataFrame(results)
    return results_df

dependent_var_list = ['MAU_숫자']  # 여기에 종속변수를 추가
independent_var_list = ['나이_숫자변환']  # 여기에 독립변수를 추가

# 결과 저장용 데이터프레임 초기화
final_results = pd.DataFrame()

# 리스트를 사용하여 분석 수행
for dependent_var in dependent_var_list:
    for independent_var in independent_var_list:
        results_df = perform_anova(df, dependent_var, independent_var)
        final_results = pd.concat([final_results, results_df], ignore_index=True)

# 결과 출력
final_results
```
